```python
import csv
import numpy as np
from matplotlib import pyplot as plt

expDat = np.genfromtxt("expDat.txt",skip_header=1,delimiter=",") 
expDat =expDat[:,1:]
x = np.arange(0,100)
for y in range(20):
    plt.plot(x,expDat[:,y])
    
fig, axs = plt.subplots(5,5,figsize=(12, 12))
for x in range(5):
    for y in range(5):
        axs[x,y].scatter(expDat[:,x],expDat[:,y])
        

fig1, (ax1,ax2) = plt.subplots(1, 2,figsize=(14, 14))
        
cov = np.cov(expDat.T)
im1 = ax1.imshow(cov)

centerMat = np.mean(expDat,axis = 0).reshape((20,1))
covcal = expDat.T-centerMat

# covariace muss divided by (nsamples - 1)
cov1 = (covcal @ covcal.T)/(expDat.shape[0]-1)
im2 = ax2.imshow(cov1)

fig1.colorbar(im1, ax=ax1,shrink=0.4)
fig1.colorbar(im2, ax=ax2,shrink=0.4)
```
